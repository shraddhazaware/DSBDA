```python
#detection of outliers
# Boxplot method--> Summaries sample data using 25th, 50th,75th value
col=['Maths_Score','Reading_Score','Writing_Score','Placement_Score']
df.boxplot(col)
```

```python
#Detection of outliers using scatterplot 

fig,ax=plt.subplots(figsize=(18,10))
ax.scatter(df['Placement_Score'],df['Placement offer count'])
plt.show()
```

```python
#Identifying Outliers with Interquartile Range (IQR) Calculate and print Quartile 1 and Quartile
q1 = np.percentile(df['Maths_Score'],25)
q3 = np.percentile(df['Maths_Score'],75)
print(q1,q3)
```

```python
#Calculate value of IQR (Inter Quartile Range)
IQR = q3-q1
#Calculate and print Upper and Lower Bound to define the outlier base value.
lwr_bound = q1-(1.5*IQR)
upr_bound = q3+(1.5*IQR)
print(lwr_bound, upr_bound)
```

```python
#printing indexes of outliers values
index_outliers = np.where((df[col] < lwr_bound) | ( df[col] > upr_bound))
index_outliers
```

```python
# handling of oulier with z score method
from scipy import stats
df2=pd.read_csv("StudentPerformance.csv")
df2
```

```python
z = np.abs(stats.zscore(df2['Maths_Score']))
print(z)
```

```python
 Upperthreshold = 1.4
Lowerthreshold = 0.18
index_outliers = np.where((z < Lowerthreshold) | ( z > Upperthreshold))
index_outliers
```

```python
#Handling of outliers using flooring and capping
#capping:outlier is capped above 90th percentile or
#floored:floored at a factor below 10th percentile
df1=df
df[col] = np.where(df1[col]< lwr_bound,lwr_bound,df[col])
df[col] = np.where(df1[col]>upr_bound ,upr_bound,df[col])
df1
```
